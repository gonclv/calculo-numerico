```python
import pandas as pd
```

```python
colunas = ['iteração', 'x', 'y', 'z', 'xn', 'yn', 'zn', 'ex', 'ey', 'ez']
arquivos = pd.DataFrame(columns = colunas)
arquivos
```

```python
def er(novo, velho):
  err = (novo - velho)/velho
  if err < 0:
    return -err
  else:
    return err
```

```python
x = 0
y = 0
z = 0
i = 0
N = 4
errox = 100
erroy = 100
erroz = 100
erro_min = 1

while i < N:
  xn = (10 - 4*z)/2
  yn = (-6 - x + z)/3
  zn = 8 - 2*x - y
  if i > 1:
    errox = er(xn, x)
    erroy = er(yn, y)
    erroz = er(zn, z)
  dados = [i, x, y, z, xn, yn, zn, errox, erroy, erroz]
  arquivos.loc[i] = dados
  x = xn
  y = yn
  z = zn
  i = i + 1
```
