```python
from pylab import *

# avalia o polinomio interpolador usando polinomios de Lagrange
def interpolL(x,T):
    S=0.0
    for i in range(len(T)):
        prod=1.0
        for j in range(len(T)):
            if j==i: continue
            prod=prod*(x-T[j][0])/(T[i][0]-T[j][0])
        S=S+T[i][1]*prod
    return S
T=[(1,2),(2,6),(3,-1)] # A tabela é uma lista de pares ordenados.
            
p= lambda x: interpolL(x,T) # define o polinomio interpolador p(x)

x=linspace(1,3,50)
plot(x,p(x))
grid()
show()
```
